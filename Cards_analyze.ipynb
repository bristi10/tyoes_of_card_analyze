```python
import pandas as pd
import numpy as np

df = pd.read_csv("cards_data.csv")

credit_cards = df[df["card_type"] == "Credit"]

# Step 2: Count credit card brands
credit_brand_count = credit_cards["card_brand"].value_counts()

# Step 3: Print result
print(credit_brand_count)

# Step 4: Find most used credit card brand
most_used_credit = credit_brand_count.idxmax()

print("Most used credit card brand:", most_used_credit)


common_card = df["card_type"].value_counts()
print("common_card:", common_card.head(1))


df["avg_credit_limit"] = df["credit_limit"].str.replace("$", "").astype(float).mean()
print("avg_credit_limit:", df["avg_credit_limit"])

max_credit_limit = df["credit_limit"].str.replace("$","").astype(float).max()
print("max_credit_limit:", max_credit_limit)

min_credit_limit = df["credit_limit"].str.replace("$", "").astype(float).min()
print("min_credi_limit:", min_credit_limit)

has_chip = df["has_chip"].value_counts()
print("has_Chip:",has_chip)

dark_web = df["card_on_dark_web"].value_counts()
print("dark_web:", dark_web)

highest_avg_credit_limit = df.groupby("card_brand")["avg_credit_limit"].idxmax()
print("highest_avg_credit_limit:", highest_avg_credit_limit)

card_issue = df.groupby("client_id")["num_cards_issued"].sum().sort_values(ascending = False)
print("card_issue:",card_issue)

total_card_issue = df["num_cards_issued"].sum()
print("total_Card_issue:",total_card_issue)

acc_open = df["acct_open_date"].value_counts()
print(acc_open.idxmax())

pin_change = df["year_pin_last_changed"].value_counts()
print(pin_change.idxmax())

df["credit_limit_card"] = df["credit_limit"].str.replace("$", "").astype(float)
print(df["credit_limit_card"])
chip_card_credit_limit= df.groupby("has_chip")["credit_limit_card"].mean()
print(chip_card_credit_limit)

dark_web = df[df["card_on_dark_web"]==True]["card_brand"].value_counts()
print(dark_web)
```
